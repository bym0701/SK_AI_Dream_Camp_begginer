```python
#연습문제
#데이터를 다시 로드 합니다.

df = sns.load_dataset("titanic")
df.head()
```

```python
#다음 조건을 만족하는 코드를 입력하세요.

#나이가 30살 이상 남자 승객 조건 필터링
#fare를 많이 낸 순서로 내림차순 정렬
#상위 10개를 출력

cond1 = (df['age']>=30)
cond2 = (df['sex']=='male')
df.loc[cond1&cond2].sort_values(by='fare', ascending=False).head(10)
```

```python
#나이가 20살 이상 40살 미만인 승객
#pclass가 1등급 혹은 2등급인 승객
#열(column)은 survived, pclass, age, fare 만 나오게 출력
#10개만 출력
cond1 = ((df['age']>=20)&(df['age']<40))
cond2 = ((df['pclass']==1)|(df['pclass']==2))
df.loc[cond1&cond2, ['survived', 'pclass', 'age', 'fare']].head(10)
```

```python
#iloc
#loc와 유사하지만, index만 허용합니다.
#loc와 마찬가지고, indexing / slicing 모두 가능합니다.
df.head()
```

```python
#indexing
df.iloc[1, 3]
```

```python
#Fancy Indexing
df.iloc[[0, 3, 4], [0, 1, 5, 6]]
```

```python
#Slicing
df.iloc[:3, :5]
```

```python
#isin
#특정 값의 포함 여부는 isin 함수를 통해 비교가 가능합니다. (파이썬의 in 키워드는 사용 불가 합니다.)
sample = pd.DataFrame({'name': ['kim', 'lee', 'park', 'choi'],
                        'age': [24, 27, 34, 19]
                      })
sample
```

```python
sample['name'].isin(['kim', 'lee'])
```

```python
sample.isin(['kim', 'lee'])
```

```python
#loc를 활용한 조건 필터링으로도 찰떡궁합입니다.
condition = sample['name'].isin(['kim', 'lee'])
```

```python
sample.loc[condition]
```

```python
#연습문제
tips = sns.load_dataset('tips')
tips.head()
```

```python
#tips 데이터셋 중 day가 금요일(Fri), 토요일(Sat) 만 필터링 합니다.
#tip이 $10보다 적게 낸 데이터만 필터링합니다.
#컬럼은 total_bill, tip, smoker, time만 출력합니다.
#상위 10개 행만 출력합니다.
cond1 = tips['day'].isin(['Fri', 'Sat'])
cond2 = tips['tip'] < 10
tips.loc[cond1 & cond2, ['total_bill', 'tip', 'smoker', 'time']].head(10)
```

```python
#나이가 30살 이상 남자 승객 조건 필터링
#fare를 많이 낸 순서로 내림차순 정렬
#상위 10개
df = sns.load_dataset("titanic")
cond = (df['age'] >= 30) & (df['sex'] == 'male')
result_df = df.loc[cond].sort_values(by='fare', ascending=False).head(10)
result_df
```
